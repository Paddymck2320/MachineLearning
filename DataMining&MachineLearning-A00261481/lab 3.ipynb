```python
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
x = np.array([1, 2, 3, 4, 5, 6]).reshape(-1, 1)
y = np.array([6,1,9,5,17,12])
print("x.shape ", x.shape)
model = LinearRegression()
model.fit(x, y)
print('intercept:', model.intercept_)
print('slope:', model.coef_)
y_hat = model.predict(x)
print('predicted response:', y_hat, sep='\n')
plt.scatter(x, y, color='black')
plt.plot(x, y_hat, color='blue', linewidth=3)
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
x = np.array([1, 2, 3, 4, 5, 6])
y = np.array([6,1,9,5,17,12])
n = len(x)
print(n)
numerator = n * np.sum(x*y) - np.sum(x)*sum(y)
print (numerator)
```

```python
denominator = n * np.sum(x*x) - np.sum(x)**2
print(denominator)
```

```python
b = numerator / denominator 
print(b)
```

```python
a = (np.sum(y) - b * np.sum(x))/n
print(a)
```

```python
import pandas as pd
import numpy as np

df = pd.read_csv('data/faithful.csv')
x = np.array(df['waiting']).reshape(-1,1)
y = np.array(df['eruptions'])
```

```python
plt.scatter(x, y, color='black', label='actual')
plt.xlabel('waiting')
plt.ylabel('eruptions')
#plt.show()
plt.savefig('plots/p2predictedActual.png')
```

```python
model = LinearRegression()
model.fit(x, y)

y_hat = model.predict(x)
plt.scatter(x, y, color='black', label='actual')
plt.scatter(x, y_hat, color='blue', label='predicted')
plt.xlabel('waiting')
plt.ylabel('eruptions')
plt.legend(loc='upper left')
# plt.show()
plt.savefig('plots/p2predictedActual.png')
```

```python
print('coefficient of determination:', model.score(x, y))
```
